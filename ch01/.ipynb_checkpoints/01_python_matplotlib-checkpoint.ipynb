```python
import numpy as np
import matplotlib.pyplot as plt

# data setting
# from 0 to 6 with 0.1 unit
x = np.arange(0, 6, 0.1)

# y = f(x) = sin(x)
y = np.sin(x)

# plot
plt.plot(x, y)
plt.show()
```

```python
# data setting
# from 0 to 6 with 0.1 unit
x = np.arange(0, 6, 0.1)

# y = f(x) = sin(x)
y1 = np.sin(x)
y2 = np.cos(x)

# plot
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")

# give name to each axis
plt.xlabel("x")
plt.ylabel("y")

# title
plt.title("sin & cos")

# show legend
plt.legend()

plt.show()
```
